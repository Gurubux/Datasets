```
import pandas as pd
import numpy as np
```

```
import seaborn as sns
import matplotlib.pyplot as plt
```

```
X_rac = pd.read_csv('https://raw.githubusercontent.com/Gurubux/Datasets/master/target_k_rac.csv')
X_cnf = pd.read_csv('https://raw.githubusercontent.com/Gurubux/Datasets/master/target_k_cnf.csv')
```

# **K_RAC**

# Correlation
### Correlation with target_k_rac in range of +/-0.2

```
plt.figure(figsize = (20,5))
X_rac.corr()['target_k_rac'].sort_values().plot.bar()
```

# Features with Negative Correlation with target_k_rac 

```
X_rac.corr()['target_k_rac'].sort_values()[:10].index.values
```

```
plt.figure(figsize = (10,10))
ax = sns.heatmap(X_rac.loc[:,np.append(X_rac.corr()['target_k_rac'].sort_values()[:10].index.values,'target_k_rac')].corr(), linewidths=.5,annot=True)
```

```
plt.figure(figsize = (10,10))
sns.pairplot(X_rac.loc[:,np.append(X_rac.corr()['target_k_rac'].sort_values()[:10].index.values,'target_k_rac')])
```

# Features with Positive Correlation with target_k_rac 

```
plt.figure(figsize = (10,10))
ax = sns.heatmap(X_rac.loc[:,X_rac.corr()['target_k_rac'].sort_values()[-10:].index.values].corr(), linewidths=.5,annot=True)
```

```
plt.figure(figsize = (10,10))
sns.pairplot(X_rac.loc[:,X_rac.corr()['target_k_rac'].sort_values()[-10:].index.values])
```
